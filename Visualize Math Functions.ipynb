```python
import numpy as np  
import math
import matplotlib.pyplot as plt  
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = eval(formula)
    plt.plot(x, y)  
    plt.show()
```

```python
graph('-1*(2*x-1)**2+1', np.arange( 0, 1,0.01))
```

```python
graph('x**0.5', np.arange( 0, 1,0.01))
```

```python
graph('21*4*(x/225)*(1-x/225)', np.arange( 0, 300,0.01))
```

```python
graph('50*4*(x/300)*(1-x/300)', np.arange( 0, 300,0.01))
```

```python
graph('100*(x/(4*143))**0.5', np.arange( 0, 5000,0.01))
```
