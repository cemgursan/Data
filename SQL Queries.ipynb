```python
# Importing necessary libs
import sqlite3
import pandas as pd
```

```python
# Connecting DB

DBconnection = sqlite3.connect("sakila.db")
```

# DB Connection Function

```python
def QuerryToDF(query):

    # Use pandas to pass sql query using connection form SQLite3
    DF = pd.read_sql(query,DBconnection)

    # Show the resulting DataFrame
    return DF
```

```python
# Write SQL query as string then pass it to SQL
# Use double " or '
# * means ALL
# after FROM you mention where we are searching the data

# This means select all customers.

# Set SQL query as a comment
query = ''' SELECT * FROM customer '''


# Use pandas to pass sql query using connection form SQLite3
# the resulting outcome is saved in DF dataframe

DF = QuerryToDF(query)
DF.head(4)
```

```python
# Let's Call last names and emails

newquery=''' SELECT last_name,email FROM customer '''

DF1=QuerryToDF(newquery)
DF1.head()
```

```python
 # columns sometimes have duplicate values
# DISTINCT can return only distinct values.
# Select distinct country_ids from the city table.

#Normal DB
query = ''' SELECT *FROM city'''
QuerryToDF(query).head()
```

```python
#DISTINCT USE
query = ''' SELECT DISTINCT(city)FROM city'''

QuerryToDF(query).head()
```

## WHERE  USE

```python
# WHERE function brings you the rows WHERE your condition is true

# Normal DB
query = '''SELECT * FROM customer'''
QuerryToDF(query).head()
```

```python
#WHERE USE with Integer

query = ''' SELECT *FROM customer WHERE customer_id = 5'''
QuerryToDF(query).head()
```

```python
#WHERE USE with String

query = ''' SELECT *FROM customer WHERE first_name = 'MARY'  '''
QuerryToDF(query).head()
```

## AND OR USE

```python
# AND OR Use

# AND Bring a row when both conditions are met
# Or Brings if one condition is met

#Normal DB
query = ''' SELECT *FROM film  '''
QuerryToDF(query).head()
```

```python
# AND USE

query = ''' SELECT *FROM film WHERE release_year = 2006 AND rental_duration =3 '''
QuerryToDF(query).head()
```

```python
# OR USE

query = ''' SELECT * FROM film WHERE rating = 'PG' OR rating = 'R' '''
QuerryToDF(query).head()
```

## COUNT() - Returns the number of rows.



```python
# Count the number of customers
query = ''' SELECT COUNT(customer_id) FROM customer; '''
QuerryToDF(query)
```

## % USE

query = ''' SELECT * FROM customer WHERE first_name LIKE 'P%' ; '''

## _ USE

```python
query = ''' SELECT *FROM customer WHERE last_name LIKE '___ANT' '''
QuerryToDF(query)
```

## GLOB USE

```python
# customers first name begins with an A or a B
query = ''' SELECT *FROM customer WHERE first_name GLOB '[AB]*' ; '''
QuerryToDF(query).head()
```

## ORDER BY

```python
# Select all customers and order results by last name
query = ''' SELECT *FROM customer ORDER BY last_name ; '''

# Use below code for descending order
# ''' SELECT *FROM customer ORDER BY last_name DESC; ''' 

QuerryToDF(query).head()
```

```python
# How many customers bought from every store!

# we select store id and count as targets
# Where? customer DB
# What do we group them by? store_id

query = ''' SELECT store_id , COUNT(customer_id) FROM customer GROUP BY store_id; '''
QuerryToDF(query)
```
