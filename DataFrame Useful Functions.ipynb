```python
# Merge

df1 = DataFrame({"sno":["A","B","C","d","e"],"Name":["Cem","Khaled","Julio","Panya","Henrique"]})
print(df1)
print("\n")
df2 =DataFrame({"sno":["A","B","C","d"],"GPA":[9,8,7,6]})
print(df2)
print("\n")

#Merge, if how= outer creates null when key is not present
#Merge, if how= inner does not create element when key is not present
#On defines the key
pd.merge(df1, df2, on="sno", how="outer") 

```

```python
#Concat


df1 = DataFrame({"sno":["A","B","C","d","e"],"Name":["Cem","Khaled","Julio","Panya","Henrique"]})
print(df1)
print("\n")
df2 =DataFrame({"sno":["A","B","C","d"],"GPA":[9,8,7,6]})
print(df2)
print("\n")

#Concat adds lists together without any interest on the key
pd.concat([df1, df2])
```
